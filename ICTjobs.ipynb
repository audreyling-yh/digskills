```python
import os
import pandas as pd
from collections import Counter
```

```python
ictjobs=pd.read_csv('data/ict_jobs/ict_jobs_ssoc.csv')
digssoc=ictjobs['ssoc4d'].tolist()
```

```python
def analyse_ssic(ssic_str):
    folder='data/jobpostings_bert_split/'

    for i in os.listdir(folder):
        filepath=folder+i
        df=pd.read_csv(filepath)
        df.drop_duplicates(inplace=True)
        
        df['ssic4d']=df['HIRE_ORG_SSIC_CODE'].apply(lambda x: str(x)[:4])
        df=df[df['ssic4d']==ssic_str]
        df=df[df['ssoc4d'].isin(digssoc)]
        
        print(Counter(df['HIRE_ORG_SSIC_CODE']))
```

```python
def analyse_ssoc(ssic_str):
    folder='data/jobpostings_bert_split/'

    for i in os.listdir(folder):
        filepath=folder+i
        df=pd.read_csv(filepath)
        df.drop_duplicates(inplace=True)
        
        df['ssic4d']=df['HIRE_ORG_SSIC_CODE'].apply(lambda x: str(x)[:4])
        df=df[df['ssic4d']==ssic_str]
        df=df[df['ssoc4d'].isin(digssoc)]
        
        print(Counter(df['ssoc4d']))
```

```python
analyse_ssic('7810')
```

```python
analyse_ssic('6109')
```

```python
analyse_ssoc('4100')
```
