```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import itertools
```

```python
df1_path='img/data/subtracks_by_year_top_ssoc1.csv'
df2_path='img/data/subtracks_by_year_top_ssoc2.csv'
df3_path='img/data/subtracks_by_year_top_ssoc3.csv'
df1=pd.read_csv(df1_path)
df2=pd.read_csv(df2_path)
df3=pd.read_csv(df3_path)
df=df1.append(df2)
df=df.append(df3)
```

```python
df
```

```python
# infrastructure
subtrack_list=['build and maintain','plan and design']
df['year']=df['year'].apply(str)
temp=df[df['subtracks_final'].isin(subtrack_list)]

sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(5,5))

sns.lineplot(data=temp, x='year', y='scaled_JOB_POST_ID', hue='subtracks_final', palette='rocket_r',ci=None)

plt.title('Subtrack demand under {}'.format('Infrastructure'))
sns.despine(top=True, right=True,left=True)

ax.legend(title='Sub-tracks',loc='upper center', bbox_to_anchor=(0.5, -0.1))

ax.set_ylim(bottom=0,top=12000)
plt.xlabel('Year')
plt.ylabel('Number of job postings')
plt.tight_layout()

plt.savefig('img/top_infrastructure_subtracks_by_year.png')
plt.close()
```

```python
ssoc1_top_subtracks=df1.groupby(['ssoc4d','subtracks_final'])['scaled_JOB_POST_ID'].sum().reset_index()
ssoc1_top_subtracks.sort_values(by=['scaled_JOB_POST_ID'],inplace=True,ascending=False)

top5_subtracks_per_ssoc1=pd.DataFrame()
for i in ssoc1_top_subtracks['ssoc4d'].unique():
    temp=ssoc1_top_subtracks[ssoc1_top_subtracks['ssoc4d']==i]
    temp=temp.head(5)
    top5_subtracks_per_ssoc1=top5_subtracks_per_ssoc1.append(temp)
temp_df1=df1.merge(top5_subtracks_per_ssoc1[['ssoc4d','subtracks_final']],on=['ssoc4d','subtracks_final'],how='inner')
    
    

ssoc2_top_subtracks=df2.groupby(['ssoc4d','subtracks_final'])['scaled_JOB_POST_ID'].sum().reset_index()
ssoc2_top_subtracks.sort_values(by=['scaled_JOB_POST_ID'],inplace=True,ascending=False)

top5_subtracks_per_ssoc2=pd.DataFrame()
for i in ssoc2_top_subtracks['ssoc4d'].unique():
    temp=ssoc2_top_subtracks[ssoc2_top_subtracks['ssoc4d']==i]
    temp=temp.head(5)
    top5_subtracks_per_ssoc2=top5_subtracks_per_ssoc2.append(temp)
temp_df2=df2.merge(top5_subtracks_per_ssoc2[['ssoc4d','subtracks_final']],on=['ssoc4d','subtracks_final'],how='inner')



ssoc3_top_subtracks=df3.groupby(['ssoc4d','subtracks_final'])['scaled_JOB_POST_ID'].sum().reset_index()
ssoc3_top_subtracks.sort_values(by=['scaled_JOB_POST_ID'],inplace=True,ascending=False)

top5_subtracks_per_ssoc3=pd.DataFrame()
for i in ssoc3_top_subtracks['ssoc4d'].unique():
    temp=ssoc3_top_subtracks[ssoc3_top_subtracks['ssoc4d']==i]
    temp=temp.head(5)
    top5_subtracks_per_ssoc3=top5_subtracks_per_ssoc3.append(temp)
temp_df3=df3.merge(top5_subtracks_per_ssoc3[['ssoc4d','subtracks_final']],on=['ssoc4d','subtracks_final'],how='inner')
```

```python
dflist=[temp_df1,temp_df2,temp_df3]
for idx, df in enumerate(dflist):
    df['year']=df['year'].apply(str)

    sns.set_style('whitegrid')

    fig, axs = plt.subplots(1, 5,figsize=(17,7))
    ssoclist=df['ssoc4d'].unique().tolist()
    subtracks=df['subtracks_final'].unique().tolist()

    g1=sns.lineplot(data=df[df['ssoc4d']==ssoclist[0]], x='year', y='scaled_JOB_POST_ID', hue='subtracks_final', palette='Set2',ax=axs[0])
    g2=sns.lineplot(data=df[df['ssoc4d']==ssoclist[1]], x='year', y='scaled_JOB_POST_ID', hue='subtracks_final', palette='Set2',ax=axs[1])
    g3=sns.lineplot(data=df[df['ssoc4d']==ssoclist[2]], x='year', y='scaled_JOB_POST_ID', hue='subtracks_final', palette='Set2',ax=axs[2])
    g4=sns.lineplot(data=df[df['ssoc4d']==ssoclist[3]], x='year', y='scaled_JOB_POST_ID', hue='subtracks_final', palette='Set2',ax=axs[3])
    g5=sns.lineplot(data=df[df['ssoc4d']==ssoclist[4]], x='year', y='scaled_JOB_POST_ID', hue='subtracks_final', palette='Set2',ax=axs[4])

    for i,ssoc in enumerate(ssoclist):
        axs[i].set_title('SSOC4D '+str(ssoclist[i]))
        axs[i].legend(title='Sub-tracks',loc='upper center', bbox_to_anchor=(0.5, -0.1))

        sns.despine(top=True, right=True,left=True)

        if i==2:
            axs[i].set_xlabel("Year")
        else:
            axs[i].set_xlabel("")

        if i==0:
            axs[i].set_ylabel("Number of job postings")
        else:
            axs[i].set_ylabel('')

    fig.suptitle('Number of job postings requiring each digital skill sub-track for SSOC Group {}'.format(str(idx+1)))

    plt.tight_layout()
    plt.savefig('img/top_ssoc{}_subtracks_by_year.png'.format(str(idx+1)))
    plt.close()
```
