```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import itertools
```

```python
df1_path='img/data/tracks_by_year_top_ssoc1.csv'
df2_path='img/data/tracks_by_year_top_ssoc2.csv'
df3_path='img/data/tracks_by_year_top_ssoc3.csv'
df1=pd.read_csv(df1_path)
df2=pd.read_csv(df2_path)
df3=pd.read_csv(df3_path)
```

```python
df1
```

```python
dflist=[df1,df2,df3]
for idx, df in enumerate(dflist):
    df['year']=df['year'].apply(str)

    sns.set_style('whitegrid')

    fig, axs = plt.subplots(1, 5,figsize=(17,4))
    ssoclist=df['ssoc4d'].unique().tolist()
    tracklist=df['tracks_final'].unique().tolist()

    g1=sns.lineplot(data=df[df['ssoc4d']==ssoclist[0]], x='year', y='scaled_JOB_POST_ID', hue='tracks_final', palette='Set2',ax=axs[0])
    g2=sns.lineplot(data=df[df['ssoc4d']==ssoclist[1]], x='year', y='scaled_JOB_POST_ID', hue='tracks_final', palette='Set2',ax=axs[1])
    g3=sns.lineplot(data=df[df['ssoc4d']==ssoclist[2]], x='year', y='scaled_JOB_POST_ID', hue='tracks_final', palette='Set2',ax=axs[2])
    g4=sns.lineplot(data=df[df['ssoc4d']==ssoclist[3]], x='year', y='scaled_JOB_POST_ID', hue='tracks_final', palette='Set2',ax=axs[3])
    g5=sns.lineplot(data=df[df['ssoc4d']==ssoclist[4]], x='year', y='scaled_JOB_POST_ID', hue='tracks_final', palette='Set2',ax=axs[4])

    g1.get_legend().remove()
    g2.get_legend().remove()
    g3.get_legend().remove()
    g4.get_legend().remove()
    g5.get_legend().remove()

    for i,ssoc in enumerate(ssoclist):
        axs[i].set_title('SSOC4D '+str(ssoclist[i]))
        sns.despine(top=True, right=True,left=True)

        if idx==0:
            axs[i].set_ylim(0,25000)
        elif idx==1:
            axs[i].set_ylim(0,30000)
        else:
            axs[i].set_ylim(0,25000)
            

        if i==2:
            axs[i].set_xlabel("Year")
        else:
            axs[i].set_xlabel("")

        if i==0:
            axs[i].set_ylabel("Number of job postings")
        else:
            axs[i].set_ylabel('')

    fig.suptitle('Number of job postings requiring each digital skill main track for SSOC Group {}'.format(str(idx+1)))

    axs.flatten()[-2].legend(title='Main track',loc='center left', bbox_to_anchor=(2.2, 0.5), ncol=1)
    plt.tight_layout()
    plt.savefig('img/top_ssoc{}_tracks_by_year.png'.format(str(idx+1)))
    plt.close()
```

```python
# delta

df=df1.append(df2).append(df3)
df["rank"] = df.groupby(['ssoc4d','year'])['scaled_JOB_POST_ID'].rank("dense", ascending=False)
df=df[df['year'].isin([2018,2021])]
df.sort_values(by=['ssoc4d','tracks_final','year'],inplace=True)
df['rankdelta']=df[['rank']].diff() # change in ranking from 2018 to 2021 (negative means rise in ranks)
df=df[df['year']==2021]
df
```
