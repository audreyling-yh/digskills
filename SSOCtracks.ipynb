```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filepath='img/data/tracks_by_top_ssoc4d_year.csv'
df=pd.read_csv(filepath)
```

```python
df
```

```python
sns.set_style('whitegrid')

fig, ax = plt.subplots(figsize=(10,5))

ax=sns.lineplot(data=df, x='year', y='scaled_JOB_POST_ID', hue='ssoc4d', style='tracks_final', palette='Set2')

sns.despine(top=True, right=True, left=True, bottom=True)
ax.set_ylim(bottom=0)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.xlabel('Year')
plt.ylabel('Number of job postings')
plt.title('Number of job postings requiring each digital skill main track by SSOC4D')
plt.tight_layout()
plt.show()
```
